```python
sns.boxplot(x=df['Temperature'])
plt.show()

print("The Median Temperature is around 21 degree celsius."
      "\nTemperature is almost low as almost half of the values are between -20 and 20 degree celsius."
      "\nDistribution is left-skewed, suggesting fewer extreme cold events compared to consistently warmer conditions."
      "\nUpper-end outliers indicate occasional heatwave conditions")
```

```python
sns.boxplot(x=df['Wind Speed'])
plt.show()

print("The Median Wind Speed is around 9 Km/h.\nThe Middle 50% of values lie between 0 to 9 Km/h. It indicates moderate variability.\nThe Median lies near the Center with nearly equal Whisker lengths, suggesting that the data is evenly spread with no strong skewness (Symmetric Distribution).\nA few high-value outliers indicate extreme wind events at some occasion")
```

```python
sns.boxplot(x=df['Precipitation (%)'])
plt.show()


print("The Median Precipitation is around 58%.\nThe Middle 50% of values lie between 0 to 59%. It indicates Above moderate Precipitation."
      "\nThe Median lies slightly right to thr center means distribution is left skewed.It suggests that high precipitation is consistent and low precipitation is not frequent.\nThere is no outlier.It tells that no extreme event happen with respect to precipitation")
```

```python
sns.boxplot(x=df['Humidity'],color="b")
plt.show()


print("Median humidity is around 70.\nHalf of the values lie between 20 to 70,suggesting comfortable humidity.\nDistribution is Symmetric (Median is almost at center) means data is evenly spread.\nThere is no outlier means no anomaly in Humidity is seen ")
```

```python
sns.boxplot(x=df['Visibility (km)'],color="b")
plt.show()

print("Median visibility is around 5 km."
      "\n50% of the values lie between o to 5 km suggesting limited or moderate visibility."
      "\nDistribution is slightly right-skewed means few instances of high visibility."
      "\nOutliers show some days have exceptionally clear visibility")
```

```python
# Bi-variate analysis

sns.boxplot(x='Weather Type', y='Temperature', data=df,color="b")
plt.show()

print("This boxplot explains distribution of Temperature across different Weather Type."
      "\nSunny Weather exhibits slightly higher median Temperature,Rainy and Cloudy Weathers have almost same median Temperature while Snowy Weather shows lower median Temperature."
      "\nSnowy Weather has slightly narrow whisker compared to other Weather types that indicates Snowy weather has lesser Temperature variability compared to other Weather types."
      "\n")
```

```python
sns.boxplot(x='Weather Type', y='Humidity', data=df,color="b")
plt.show()

print("This explains distribution of Humidity across different Weather Type."
      "\nRainy and Snowy Weather have more humidity than other weathers."
      "\nSunny Weather has wider whisker compared to other Weather types that shows more humidity variation in Sunny weather."
      "\nThe outliers in Rainy and Snowy Weathers show low humidity days sometimes."
      "\n ")
```

```python
sns.boxplot(x='Weather Type', y='Precipitation (%)', data=df,color="b")
plt.show()

print("Rainy and Snowy Weather have higher median precipitation than other weathers."
      "\nOutliers in Sunny weather shows exceptionally precipitation in Sunny days also."
      "\n")
```

```python
sns.boxplot(x='Weather Type', y='Wind Speed', data=df,color="b")
plt.show()
print("Rainy weather has slightly higher wind speed than other weathers."
      "\nSnowy weather has higher variation in wind speed than other weathers as it has wider whisker."
      "\nOutliers in Rainy and Snowy Weathers show very high wind speed at some days.")
```

```python
sns.boxplot(x='Weather Type', y='Visibility (km)', data=df,color="b")
plt.show()

print("Rainy and Snowy Weather have lower visibility than other weathers."
      "\nSunny weather has higher visibility than other weathers."
      "\nOutliers in Rainy amd Snowy weathers show that there are some days when visibility is very high and clear")
```

```python
num_df = df.select_dtypes(include="number")
corr = num_df.corr()
print(corr)
```

```python
plt.figure(figsize = (10,6))
sns.heatmap(corr, annot=True, cmap="YlGnBu", fmt="2f")
plt.title("Correlation Heatmap of weather variables")
plt.show()
```

```python
weather_mean = df.groupby('Weather Type')["Humidity"].mean()
print(weather_mean)
```
